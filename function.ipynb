# 指数信号

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
A = 1
w0 = 2 * np.pi
phi = np.pi / 6

# 定义时间轴：和Matlab的 t=0:0.001:8 对应
t = np.linspace(0, 8, 8001)  # 8001个点对应步长0.001

# 计算信号
x2 = A * np.cos(w0 * t + phi)

# 绘图
plt.xlabel('t')
plt.ylabel('x2(t)')
plt.plot(t, x2)
plt.axis([0, 8, -1, 1])  # 设置坐标轴范围
plt.show()
```

# 抽样信号

例1-5

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义n范围：-5到5
n = np.arange(-5, 6)
# 单位冲激：n=0时为1，其余为0
delta = np.where(n == 0, 1, 0)
plt.stem(n, delta)
plt.show()
```

# 连续时间信号组合

例1-10

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-20, 20, 1000)

x = np.where(t < -2, 0,
            np.where((t >= -2) & (t < 0), -4 - 2*t,
                    np.where((t >= 0) & (t < 4), -4 + 3*t,
                            np.where((t >= 4) & (t < 8), 16 - 2*t, 0)
                            )
                    )
            )

plt.plot(t, x)
plt.grid(True) # 加网格更清晰
plt.show()
```
